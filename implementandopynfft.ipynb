```python
import numpy as np
from pynfft.nfft import NFFT

#comentario teste de token
def funcao_nfft(x, f_hat):
    N = [16, 16]  # Domínio da frequência (2D)
    M = len(x)        # Número de amostras
    plan = NFFT(N, M)
    plan.x = x
    plan.precompute()  # Pré-computação

    plan.f_hat = f_hat
    # Transformada direta: f_hat -> f
    f = plan.trafo()

    # Transformada adjunta: f -> f_hat
    plan.f = f
    return f
```

```python


# Inicializa o plano NFFT
N = [16, 16]  # Domínio da frequência (2D)
M = 96        # Número de amostras
plan = NFFT(N, M)

# Define as coordenadas no domínio [-0.5, 0.5]^2
x = np.random.rand(M, 2) - 0.5  # Coordenadas bidimensionais
plan.x = x
plan.precompute()  # Pré-computação

# Define os dados no domínio da frequência (f_hat)
f_hat = np.random.rand(N[0], N[1]) + 1j * np.random.rand(N[0], N[1])  # Dados complexos
plan.f_hat = f_hat

# Transformada direta: f_hat -> f
f = plan.trafo()
print("Transformada Direta (f):", f)

# Transformada adjunta: f -> f_hat
plan.f = f
f_hat_reconstructed = plan.adjoint()
#print("Reconstrução Adjunta (f_hat):", f_hat_reconstructed)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pynfft.nfft import NFFT

# Configuração dos parâmetros
N = [16, 16]  # Domínio da frequência (2D)
M = 96        # Número de amostras
plan = NFFT(N, M)

# Coordenadas no domínio [-0.5, 0.5]^2
x = np.random.rand(M, 2) - 0.5  # Coordenadas bidimensionais
plan.x = x
plan.precompute()  # Pré-computação

# Plotando o espaço amostral
plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], color='blue', label='Amostras')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlim([-0.5, 0.5])
plt.ylim([-0.5, 0.5])
plt.title("Espaço Amostral 2D")
plt.xlabel("Coordenada x")
plt.ylabel("Coordenada y")
plt.legend()
plt.grid()
plt.show()
```

```python
# Gerando valores associados (complexos)
f = np.random.rand(M) + 1j * np.random.rand(M)

# Magnitude dos valores para visualização
magnitude = np.abs(f)

# Plotando as magnitudes no espaço amostral
plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], c=magnitude, cmap='viridis', label='Magnitude')
plt.colorbar(label="Magnitude de f")
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlim([-0.5, 0.5])
plt.ylim([-0.5, 0.5])
plt.title("Valores Associados às Amostras")
plt.xlabel("Coordenada x")
plt.ylabel("Coordenada y")
plt.legend()
plt.grid()
plt.show()
```
